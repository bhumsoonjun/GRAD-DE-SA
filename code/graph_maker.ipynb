```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
test_names = [
    ("sphere", "Sphere"),
    ("weighted_sphere", "Weighted Sphere"),
    ("schwefel_1_2", "Schwefel 1.2"),
    ("schwefel_2_3", "Schwefel 2.3"),
    ("rotated_hyper_ellipsoid", "Rotated Hyper Ellipsoid"),
    ("rosenbrock", "Rosenbrock"),
    ("pow_sum", "Sum of Different Powers"),
    ("easom", "Easom"),
    ("ackley", "Ackley"),
    ("rastrigin", "Rastrigin"),
    ("griewangk", "Griewangk"),
    ("dejong_5", "De Jong 5"),
    ("dejong_3", "De Jong 3")
]
algo_names = [
    ("ttp_de_sa_2", "DE-SA"),
    ("ge", "GE"),
    ("de", "DE"),
    ("da", "DA"),
    ("abc", "ABC"),
    ("pso", "PSO")
]
```

```python
def load_cost_array_as_numpy(path: str):
    return np.load(path)
```

```python
for test_label, test_name in test_names:
    # plot lineplot for each algo on same graph
    # log scale y axis
    plt.figure()
    plt.yscale("log")
    plt.xlabel("Iteration")
    plt.ylabel("Cost")

    # black border
    plt.rcParams["axes.edgecolor"] = "black"
    # increase border width
    plt.rcParams["axes.linewidth"] = 1.5
    plt.rcParams["axes.facecolor"] = "None"

    # no need to increase font size for legend
    plt.rcParams["legend.fontsize"] = 10
    # increase font size for axes only
    plt.rcParams["axes.labelsize"] = 16
    # increas font size for ticks only
    plt.rcParams["xtick.labelsize"] = 14
    plt.rcParams["ytick.labelsize"] = 14
    # bold font for axes
    plt.rcParams["axes.labelweight"] = "bold"

    marker_style = ["o", "v", "s", "P", "X", "D"]

    # increase plot size
    plt.rcParams["figure.figsize"] = (8, 6)

    for algo_label, algo_name in algo_names:
        cost_array = load_cost_array_as_numpy(f"results/{algo_label}/{test_label}/cost.npy")
        # plot lineplot every 50th iteration with different marker
        # plot line every 50th iteration
        # crease line width
        # do not fill markers

        sns.lineplot(
            x=np.arange(0, cost_array.shape[0], 50),
            y=cost_array[::50],
            label=algo_name,
            marker=marker_style.pop(0),
            color="black",
            linewidth=0.7,
            markerfacecolor="None",
            markeredgecolor="black",
        )

    plt.legend()
    plt.savefig(f"results/{test_label}.png")
```
